```python
# NOTE: ADDS THE LOGO TO EACH IMAGE

import PIL
from PIL import Image
import os
import sys
from IPython.display import Image


right = 0
left = 0
top = 0
bottom = 0

directory = r"C:\Users\s.georgiou\AppData\Roaming\jupyter\ORDERLION png\png default\\" # the primary directory where the pictures are located
logo_directory = r"C:\Users\s.georgiou\AppData\Roaming\jupyter\ORDERLION png\logo\Reichenau Gemüse Logo Transparent.png" # the directory in which the desired logo is located


for filename in os.listdir(directory): # loop through chosen directory
    if filename.endswith(".png"): # if the filename ends with .png
        f, e = os.path.splitext(filename)
        outfile = f + ".png"
        width = 0
        height = 0
        thumb_width = 0
        thumb_height = 0
        with PIL.Image.open(f"{directory}{filename}") as im:
            logo = PIL.Image.open(f"{logo_directory}")
            im = im.convert("RGBA") # convert to RGBA color space 
            logo = logo.convert("RGBA")
            
            width = im.size[0]
            height = im.size[1]

            im2 = PIL.Image.new("RGBA", (width,int(height/3)), (255,255,255,255))

            thumb_width = int(im2.size[0]-20)
            thumb_height = int(im2.size[1]-15)

            max_thumbsize = [thumb_width,thumb_height]
            logo.thumbnail(max_thumbsize, PIL.Image.ANTIALIAS)
            print(logo.size)
            #print("imgsize:", im.size)
            #print("logosize:", logo.size)
            #print("whitspacesize:", im2.size)
            logo_paste_width, logo_paste_height = [int(width/2)-int(logo.size[0]/2),int(height)-logo.size[1]-((int(im2.size[1])-int(logo.size[1]))//int(2))] 
            # x position divided by two minus half the logo size will place the logo in frame relative to x.
            # starting height y minus logo size y subtracted by the image2 size y and the logo size y divded by 2 will place the logo in the y center relative to the semi-transparent surface (im2)
            print(logo_paste_height)

            maskim = PIL.Image.new("RGBA", (im2.size[0], int(im2.size[1])), (255,255,255,128))
            #print(padim_width,padim_height)
            
            im.paste(im=im2, box=(0,height-int(height/3)),mask=maskim)
            #padim.paste(logo, (padim_width-max_thumbsize[0]//2, padim_height-max_thumbsize[1]//2), logo)
            im.paste(logo, (logo_paste_width,logo_paste_height), logo)
            im.save(f"C:\\Users\\s.georgiou\\AppData\\Roaming\\jupyter\\ORDERLION png\\png default\\{f}.png")

```
